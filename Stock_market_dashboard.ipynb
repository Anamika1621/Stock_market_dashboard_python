```python
import yfinance as yf
import pandas as pd
import plotly.graph_objs as go
import plotly.express as px
import streamlit as st
```

```python
#fetch stock data
tesla = yf.Ticker('TSLA')
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```python
#plothistoricalchart
ticker = 'TSLA'
fig = go.Figure(data=[go.Candlestick(
    x=tesla_data['Date'],
    open=tesla_data['Open'],
    high=tesla_data['High'],
    low=tesla_data['Low'],
    close=tesla_data['Close']
)])
fig.update_layout(title=f"{ticker} Stock Price (6 months)", xaxis_rangeslider_visible=True)
fig.show()
```

```python
#Add more average
tesla_data["MA20"] = tesla_data["Close"].rolling(20).mean()
tesla_data["MA50"] = tesla_data["Close"].rolling(50).mean()

fig = px.line(tesla_data, x="Date", y=["Close", "MA20", "MA50"],
              title=f"{ticker} Closing Price with Moving Averages")
fig.show()
```

```python
%%writefile app.py
import yfinance as yf
import pandas as pd
import plotly.graph_objs as go
import plotly.express as px
import streamlit as st

st.title(" Stock Market Dashboard")

# Sidebar input
ticker = st.sidebar.text_input("Enter Stock Ticker", "AAPL")
period = st.sidebar.selectbox("Select Period", ["1mo","3mo","6mo","1y","2y","5y","10y","max"])
interval = st.sidebar.selectbox("Select Interval", ["1d","5d","1wk","1mo","3mo"])

# Fetch data
stock = yf.Ticker(ticker)
df = stock.history(period=period, interval=interval).reset_index()

# Show data
st.subheader(f"Stock Data for {ticker}")
st.write(df.tail())

# Line Chart
fig = px.line(df, x="Date", y="Close", title=f"{ticker} Closing Price")
st.plotly_chart(fig)

# Candlestick
fig2 = go.Figure(data=[go.Candlestick(
    x=df['Date'],
    open=df['Open'],
    high=df['High'],
    low=df['Low'],
    close=df['Close']
)])
fig2.update_layout(title=f"{ticker} Candlestick Chart", xaxis_rangeslider_visible=True)
st.plotly_chart(fig2)
```
